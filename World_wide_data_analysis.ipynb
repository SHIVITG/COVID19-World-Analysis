```python
list_of_countries =  list(latest_data['Country_Region'].unique())
```

```python
len(list_of_countries)
```

```python
country_specific_confirmed_cases = []
country_specific_death_cases = [] 
country_specific_recovery_cases = []
country_specific_mortality_rate = [] 

no_of_cases = []

for i in list_of_countries:
    cases = latest_data[latest_data['Country_Region']==i]['Confirmed'].sum()
    if cases > 0:
        country_specific_confirmed_cases.append(cases)
    else:
        no_of_cases.append(i)
        
for i in no_of_cases:
    list_of_countries.remove(i)
    
list_of_countries = [k for k, v in sorted(zip(list_of_countries, country_specific_confirmed_cases), key=operator.itemgetter(1), reverse=True)]
for i in range(len(list_of_countries)):
    country_specific_confirmed_cases[i] = latest_data[latest_data['Country_Region']==list_of_countries[i]]['Confirmed'].sum()
    country_specific_death_cases.append(latest_data[latest_data['Country_Region']==list_of_countries[i]]['Deaths'].sum())
    country_specific_recovery_cases.append(latest_data[latest_data['Country_Region']==list_of_countries[i]]['Recovered'].sum())
    country_specific_mortality_rate.append(country_specific_death_cases[i]/country_specific_confirmed_cases[i])
```

```python
from  geopy.geocoders import Nominatim
country_lat = []
country_long =[]
def fetch_geocodes(text):
    geolocator = Nominatim()
    loc = geolocator.geocode(text)
    lat = loc.latitude
    long = loc.longitude
#     print("latitude is :-" ,loc.latitude,",longtitude is:-" ,loc.longitude, ". Location:",text)
#     print(text)
    country_lat.append(lat)
    country_long.append(long)
```

```python
df = pd.DataFrame({'Country Name': list_of_countries, 'Number of Confirmed Cases': country_specific_confirmed_cases,
                          'Number of Deaths': country_specific_death_cases, 'Number of Recoveries' : country_specific_recovery_cases,
                          'Mortality Rate': country_specific_mortality_rate})
```
